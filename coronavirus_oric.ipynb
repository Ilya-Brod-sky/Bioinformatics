Теперь попробуем проанализировать этот участок, выяснив, есть ли там часто повторяющиеся 8- и 9-меры

```python
def frequent_words(text, k, distance):
    words = []
    freq = frequency_map(text, k, distance)
    m = max(freq.values())
    for key in freq.keys():
        if freq[key] >= m:
            words.append(key)
    return words, m

def frequency_map(text, k, distance):
    freq = {}
    n = len(text)
    for i in range(n - k + 1):
        pattern = text[i:i + k]
        if pattern not in freq.keys():
            freq[pattern] = pattern_count(text, pattern, distance)
#             freq[pattern] = reverse_complementary_pattern_count(text, pattern, distance)
    return freq

def hamming_distance(word1, word2):
    dist = 0
    for i in range(len(word1)):
        if word1[i] != word2[i]:
            dist += 1
    return dist

def pattern_count(text, pattern, d):
    count = 0
    l = len(text)
    k = len(pattern)
    for i in range(l-k+1):
        if hamming_distance(text[i:i+k], pattern) <= d:
            count += 1
    return count      
```

```python
fw90 = frequent_words(oric, 9, 0)
print("9-mers, d = 0: ", fw90)
fw80 = frequent_words(oric, 8, 0)
print("8-mers, d = 0: ", fw80)
```

Видно, что это примерно один и тот же 9-мер, однако точных совпадений не очень много, попробуем поискать ещё раз, разрешив разницу в одном символе

```python
fw91 = frequent_words(oric, 9, 1)
print("9-mers, d = 1: ", fw91)
fw81 = frequent_words(oric, 8, 1)
print("8-mers, d = 1: ", fw81)
```

Количество повторений 8- и 9-меров увеличилось до трёх, однако 9-мер поменялся. Теперь добавим reverse complementary patterns:

```python
def frequency_map(text, k, distance):
    freq = {}
    n = len(text)
    for i in range(n - k + 1):
        pattern = text[i:i + k]
        if pattern not in freq.keys():
#             freq[pattern] = pattern_count(text, pattern, distance)
            freq[pattern] = reverse_complementary_pattern_count(text, pattern, distance)
    return freq

def reverse_complement(pattern):
    return reverse(complement(pattern))

def reverse(pattern):
    reverse_pattern = ""
    for letter in pattern:
        reverse_pattern = letter + reverse_pattern
    return reverse_pattern

def complement(pattern):
    complement_pattern = ""
    for letter in pattern:
        if letter == "A":
            complement_pattern += "T"
        elif letter == "T":
            complement_pattern += "A"
        elif letter == "G":
            complement_pattern += "C"
        elif letter == "C":
            complement_pattern += "G"
    return complement_pattern

def reverse_complementary_pattern_count(text, pattern, d):
    rc_pattern = reverse_complement(pattern)
    count = 0
    l = len(text)
    k = len(pattern)
    for i in range(l-k+1):
        if hamming_distance(text[i:i+k], pattern) <= d or hamming_distance(text[i:i+k], rc_pattern) <= d:
            count += 1
    return count

fw91r = frequent_words(oric, 9, 1)
# fw81 = frequent_words(oric, 8, 1)
print("rc 9-mers, d = 1: ", fw91r)
```

Количество повторений не изменилось (что меня удивило).
